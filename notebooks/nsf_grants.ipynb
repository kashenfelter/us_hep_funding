```python
import pandas as pd
import numpy as np
```

```python
data2017 = pd.read_csv('data/nsf_grants/datafeeds\\2017_NSF_Grants_Full_20170215.csv')
data2017.columns
```

```python
data2016 = pd.read_csv('data/nsf_grants/datafeeds\\2016_NSF_Grants_Full_20161215.csv')
```

```python
data2015 = pd.read_csv('data/nsf_grants/datafeeds\\2015_NSF_Grants_Full_20151115.csv')
data2014 = pd.read_csv('data/nsf_grants/datafeeds\\2014_NSF_Grants_Full_20150515.csv')
data2013 = pd.read_csv('data/nsf_grants/datafeeds\\2013_NSF_Grants_Full_20150515.csv')
data2012 = pd.read_csv('data/nsf_grants/datafeeds\\2012_NSF_Grants_Full_20150515.csv')
```

```python
data2012.columns
```

```python
fulldata = pd.concat([data2017,data2016,data2015,data2014,data2013,data2012])
```

```python
fulldata['cfda_program_title'].drop_duplicates()
```

```python
mps_grants = fulldata[fulldata['cfda_program_title'] == 'Mathematical and Physical Sciences']
len(mps_grants)
```

```python
mps_grants = mps_grants.dropna(subset=['principal_place_cd'])
```

```python
strlist = []
for code in mps_grants['principal_place_cd'].values:
    if code == 'ZZ': code = '00'
    if len(str(int(code))) < 2:
        strlist.append('0'+str(int(code)))
    else:
        strlist.append(str(int(code)))
print strlist
```

```python
#mps_grants['fed_funding_amount']
#mps_grants['principal_place_state']
mps_grants['cong_dist'] = mps_grants['principal_place_state_code']+strlist
```

```python
pd.to_pickle(mps_grants,'nsf_mps_grants')
```
